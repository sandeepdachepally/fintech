```python

def get_rules(output_rules = pd.DataFrame()):
    li_sets = [set1, set2, set3, set4, set5, set6, set7, set8, set9, set10, set11, set12 ]
    for s in li_sets:
        te = TransactionEncoder()
        te_ary = te.fit(s).transform(s)
        df = pd.DataFrame(te_ary, columns=te.columns_)
        frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
        ass_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.25)
        ass_rules["antecedent_len"] = ass_rules["antecedents"].apply(lambda x: len(x))
        ass_rules = ass_rules[ (ass_rules['antecedent_len']<=2)]
        output_rules = pd.concat([ output_rules, ass_rules], axis = 0, ignore_index= True)
    return output_rules
    #print ("Recommendations",ass_rules)
```

```python
output_rules = get_rules()
output_rules
```

```python
def convert_list(x):
    x = list(x)
    res = ""
    for i in x:
        res = res + i +"," 
    res = res[:-1]
    return res
```

```python
output_rules["antecedents"] = output_rules["antecedents"].apply(convert_list)
output_rules["consequents"] = output_rules["consequents"].apply(convert_list)
type(output_rules.at[0,"consequents"])
```
