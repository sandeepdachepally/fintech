```python
offers_df = pd.read_csv("offers.csv")
bought_df = pd.read_csv("brandWise.csv")
```

```python
offers_df.drop(["antecedent support", "leverage", "confidence", "consequent support", "conviction", "antecedent_len", "support"],axis = 1,inplace= True)
offers_df.drop(offers_df.columns[offers_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
bought_brands = bought_df["brand"].tolist()
```

```python
offers_df
```

```python
print(type(offers_df.at[0,"antecedents"]))
```

```python
to_rec = []
for i in range(offers_df.shape[0]):
    a = offers_df.at[i,"antecedents"].split(',')
    b = offers_df.at[i,"consequents"].split(',')
    #print (a,b)
    for i in a:
        print((i).strip())
        if i in bought_brands:
            print("yes")
            for j in b:
                to_rec.append(j)    
```

```python
to_rec = set(to_rec)
```

```python
results = []
for i in to_rec:
    if i not in bought_brands:
        results.append(i)
```

```python
results
```

```python
results.remove('Allen Solly')
results.remove('Clarks')
results.remove('Tatasky')
results.remove('Patym Mall')
results.remove('Ginger')
results.remove('Fab India')
```
