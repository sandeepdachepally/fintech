```python
already_list1 = []
already_list2 = []
def random_number_id(length):
    key = ''
    for i in range(length):
        key += random.choice(string.digits)
    if key in already_list1:
        random_number_id(length)
    already_list1.append(key)
    return key
def transaction_id():
    idd = "M"+random_number_id(8)
    if idd in already_list2:
           transaction_id()
    already_list2.append(idd)
    return idd
def randomTime():
    # generate random number scaled to number of seconds in a day
    # (24*60*60) = 86,400

    rtime = int(random.random()*86400)

    hours   = int(rtime/3600)
    minutes = int((rtime - hours*3600)/60)
    seconds = rtime - hours*3600 - minutes*60

    time_string = '%02d:%02d:%02d' % (hours, minutes, seconds)
    return time_string
```

```python
def json2csv(json_name, csv_name, key):
    with open(json_name) as json_file:
        data = json.load(json_file)
    transaction_data = data[key]
    data_file = open(csv_name, 'w') 
    csv_writer = csv.writer(data_file)
    count = 0
    for tran in transaction_data:
        if count == 0: 
            # Writing headers of CSV file 
            header = tran.keys()
            csv_writer.writerow(header) 
            count += 1
        # Writing data of CSV file 
        csv_writer.writerow(tran.values())
    data_file.close() 
```

```python
def upi(b):
    return "UPI/"+random_number_id(12)+"/order/"+b+"/"+random.choice(["okaxis", "ybl", "upi", "okicici", 'okhdfcbank', "oksbi"])
```

```python
def vin(b):
    return "VIN/"+b+"/"+random_number_id(12)+"/"+random_number_id(12)
```

```python
def ips(b):
    return "IPS/"+b+"/"+random_number_id(12)+"/"+random_number_id(12)+"/"+"BANGALORE"
```

```python
def iin(b):
    return "IIN/"+"I-Debit/"+b+"/"+random_number_id(12)+"/"+random_number_id(12)
```

```python
def vps(b):
    return "VPS/"+b+"/"+random_number_id(12)+"/"+random_number_id(12)+"/"+random.choice(["BLR","BANGALOR","BANGALORE","bangalore","Begaluru","BENGALURU"])
```

```python
def mmt(b):
    return "MMT/"+"IMPS/"+random_number_id(12)+"/"+b+"/"+"axis/"+random.choice(["SBIN","HDF","SIB","KIN","BLR"])
```

```python
def atm(b,da): #date also required as arguement
    return "ATM/"+''.join(random.choices(string.ascii_uppercase , k=5))+''.join(random.choices(string.digits , k=3))+"/"+"CASH WDL/"+str(da)
```

```python
def ach(b):
    return "ACH/"+b+"/"+random_number_id(9)+"/"+random_number_id(10)
```

```python
def bil(b):
    x = ""
    if b == "Motilal Oswal":
        x = "MODIRECT"
    elif b == "Principal Mutual Fund":
        x = "PMF_RIC"
    else:
        x = random_number_id(10)
    return "BIL/"+random.choice(["ONL","BPAY","INF"])+"/"+random_number_id(12)+"/"+b+"/"+x
```

```python
def other(b, da, mo):
    return b+" "+mo+" "+da[:4]
```

```python
mode_dict = {"VIN":vin, "IPS":ips, "UPI":upi, "IIN":iin, "VPS":vps, "MMT":mmt, "ACH":ach, "BIL":bil, "ATM": atm, "OTHER":other} 
#Function to remove whitespace (called by simply_string)
def remove(string): 
    return string.replace(" ", "") 
#Function to bring entire string to lower case and remove any spaces
def simplify_string(string):
    string = remove(string)
    string = string.casefold()
    return string
```

```python
transactionTimestamp, txnId, mode, amount, balance, _type, narration, reference, valueDate = [], [], [], [], [], [], [], [], []
for index_t, row_t in df.iterrows():
            txnId1 = transaction_id() #txnId
            brand1 = row_t['brand']
            brand1 = simplify_string(brand1)
            mode1 = row_t["mode"]
            valueDate1 = row_t["valueDate"] #valueDate
            if mode1 == "ATM":
                narration1 = mode_dict[mode1](brand1, valueDate1) #narration
            elif mode1 == "OTHER":
                narration1 = mode_dict[mode1](brand1, valueDate1, row_t["month"]) #narration
            else:
                 narration1 = mode_dict[mode1](brand1) #narration
            if mode1 != "ATM":
                mode1 = "FT"  #Mode
            _type1 = row_t["type"] #type
            amount1 = str(row_t["amount"]) #amount
            reference1 = "RFN" + random_number_id(8) #reference 
            balance1 = "" #balance
            transactionTimestamp1 = str(valueDate1)+"T0"+str(randomTime())+".000Z" #transctionTimestamp
            transactionTimestamp.append(transactionTimestamp1)
            txnId.append(txnId1)
            mode.append(mode1)
            amount.append(amount1)
            balance.append(balance1)
            _type.append(_type1)
            narration.append(narration1)
            reference.append(reference1)
            valueDate.append(valueDate1)         
```

```python
output_df1 = pd.DataFrame({"transactionTimestamp":transactionTimestamp,
                "txnId":txnId,
                "mode":mode,
                "amount":amount,
                "balance":balance,
                "type":_type,
                "narration":narration,
                "reference":reference,
                "valueDate":valueDate})
```

```python
output_df1["balance"] = 0
def to_str(x):
    return str(x)
output_df1.at[0, "balance"] = 539769
```

```python
prev,current = 539769,0
for i in range(output_df1.shape[0]):
    current = ((output_df1.at[i,"balance"]))
    if current == 0:
        if output_df1.at[i,"type"]=="DEBIT":
            current = prev - int(output_df1.at[i,"amount"])
        else:
            current = prev + int(output_df1.at[i,"amount"])
        #current = str(current)
        output_df1.at[i,"balance"] = current
        prev = current
```
